```python
# Importing libraries
import json
import time
from telnetlib import Telnet
import serial
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# Initializing keys, PORT and telnet configuration

ser = serial.Serial('COM3', 9600, timeout=1);
tn=Telnet('localhost',13854);

start=time.clock();

i=0;
tn.write(b'{"enableRawOutput": true, "format": "Json"}');

eSenseDict={'attention':0, 'meditation':0};
waveDict={'lowGamma':0, 'highGamma':0, 'highAlpha':0, 'delta':0, 'highBeta':0, 'lowAlpha':0, 'lowBeta':0, 'theta':0};
signalLevel=0;
signalValue = "";
index= 1;
middle= 1;
ring= 1;
pinky= 1;
thumb= 1;
close = 1;
def hzconvert(s):
    return s[-5:2]
```

```python
EEGDATAFRAME = pd.DataFrame(columns=['Time', 'Attention', 'Meditation','Blink_Strength'])
WAVE_EEGDATAFRAME = pd.DataFrame(columns=['Time', 'Beta', 'Gamma','Alpha','Delta','Theta'])

while i<100:
    eeg_data = []
    eeg_data_wave = []

    blinkStrength=0;

    line=tn.read_until(b'\r');
    
    if len(line) > 20:
        timediff=time.clock()-start;
        dict=json.loads((line));
#         print(line);
        if "poorSignalLevel" in dict:
            signalLevel=dict['poorSignalLevel'];
        if "blinkStrength" in dict:
            blinkStrength=dict['blinkStrength'];
        if "eegPower" in dict:
            waveDict=dict['eegPower'];
#             print("waveDict",waveDict);
            eSenseDict=dict['eSense'];

        attentionStr=str(eSenseDict['attention']);
        meditationStr=str(eSenseDict['meditation']);
        betaStr=hzconvert(str(waveDict['highBeta']));
        alphaStr=hzconvert(str(waveDict['highAlpha']));
        thetaStr=hzconvert(str(waveDict['theta']));
        deltaStr=hzconvert(str(waveDict['delta']));
        gammaStr=hzconvert(str(waveDict['highGamma']));
        blinkStrengthStr=str(blinkStrength);
        seconds = time.time();
#         print(attentionStr,"attentionStr");

        eeg_data.append({'Time': seconds,'Attention': attentionStr, 'Meditation': meditationStr, 'Blink_Strength':blinkStrengthStr})
        eeg_data_wave.append({'Time': seconds, 'Beta': betaStr,'Gamma': gammaStr,'Alpha': alphaStr, 'Delta':deltaStr, 'Theta':thetaStr})    
#         print(eeg_data);
        eeg_df = pd.DataFrame(eeg_data)
        eeg_wave_df = pd.DataFrame(eeg_data_wave)
        EEGDATAFRAME = EEGDATAFRAME.append(eeg_df, ignore_index=True)
        WAVE_EEGDATAFRAME = WAVE_EEGDATAFRAME.append(eeg_wave_df, ignore_index=True)
#         print(EEGDATAFRAME) 
        EEGDATAFRAME.to_csv('eegdata.csv',index=True)
        WAVE_EEGDATAFRAME.to_csv('wave_eegdata.csv',index=True)
     
    
        if(signalLevel==200):
            signalValue = "Poor";
        else:
            signalValue = "Strong";
        

        print("Attention Value %s and Meditation Value %s --> Outpul Signal Level %s" %(attentionStr, meditationStr, signalValue))
        attentionInt = int(attentionStr)
        meditationInt = int(meditationStr)
        
        if ((attentionInt > 20) and (attentionInt < 45)):
            index += 1
            if(index==5):
                print("Move Index Finger");
                ser.write(str.encode('a'));
                index = 1;
                
            else:
                ser.write(str.encode('f'));
            
        if ((attentionInt > 45) and (attentionInt < 70)):
            middle += 1
            if(middle==5):
                print("Move Middle Finger");
                ser.write(str.encode('b'));
                middle = 1;
            else:
                ser.write(str.encode('f'));
            
        if ((attentionInt > 70) and (attentionInt < 95)):
            ring += 1
            if(ring==5):
                print("Move Ring Finger");
                ser.write(str.encode('c'));
                ring = 1;
                
            else:
                ser.write(str.encode('f'));
                

        if ((meditationInt > 20) and (meditationInt < 45)):
            pinky += 1
            if(pinky==5):
                print("Move Pinky Finger");
                ser.write(str.encode('d'));
                pinky = 1;
                
            else:
                ser.write(str.encode('f'));
            
        if ((meditationInt > 45) and (meditationInt < 70)):
            thumb += 1
            if(thumb==5):
                print("Move Thumb Finger");
                ser.write(str.encode('e'));
                thumb = 1;
                
            else:
                ser.write(str.encode('f'));
                
        
        if ((meditationInt > 70) and (meditationInt < 95)):
            close += 1
            if(close==5):
                print("Closing all fingers");
                ser.write(str.encode('f'));
                close = 1; 
            
        else:
            print("Closing all fingers");
            ser.write(str.encode('f'));
            time.sleep(1)
            
ser.close();
tn.close();
```
