```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib import style
import pandas as pd
```

```python
# plt.style.use('fivethirtyeight')
plt.style.use('seaborn')

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
```

```python
def animation(i):
    WAVE_EEGDATAFRAME = pd.read_csv('wave_eegdata.csv')
    x = []
    y = []
    
    x = WAVE_EEGDATAFRAME[0:i]['Time']
    y = WAVE_EEGDATAFRAME[0:i]['Beta']
    
    ax.clear()
    ax.plot(x, y, ax=ax, )
```

```python
animation = FuncAnimation(fig, func=animation, interval=1000)
ax.set_xlabel("Time")
ax.set_ylabel("Beta Hz Value")
plt.title("Graph for Beta Vlues")
plt.show()
```
